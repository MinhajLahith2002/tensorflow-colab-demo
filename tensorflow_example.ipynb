# import Data

```python
import pandas as pd
url="https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/Ionosphere.csv"
df=pd.read_csv(url)
df.head()
```

```python
x=df.drop(['Class'], axis=1)
y=df['Class']
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
pip list
```

# import Dependencies


```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

# Build Model

```python
model=Sequential()
model.add(Dense(units=10,activation='relu',input_dim=len(x_train.columns)))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32)
```

```python
model.save('weight.h5')
```

```python
from keras.models import load_model
model = load_model('weight.h5')

# Pick one test example
sample = x_test.iloc[0].values.reshape(1, -1)
prediction = model.predict(sample)

if prediction >= 0.5:
    print("Good signal")
else:
    print("Bad signal")


```

```python
# Predict on the whole test set
predictions = model.predict(x_test)

# Convert probabilities into Good/Bad labels
labels = ["Good" if p >= 0.5 else "Bad" for p in predictions]

# Show first 10 predictions with actual answers
for i in range(10):
    print(f"Predicted: {labels[i]} | Actual: {y_test.iloc[i]}")
```

```python
from sklearn.metrics import accuracy_score

# Convert probabilities to labels
predictions = model.predict(x_test)
labels = [1 if p >= 0.5 else 0 for p in predictions]

# Compare with actual test labels
accuracy = accuracy_score(y_test, labels)
print("Overall Test Accuracy:", accuracy)
```
